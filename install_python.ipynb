# Install of Python

Note that you can install python in many ways,
However, this is something I use daily due to the convenience!

The easiest way to get everything setup as you need to is to use a package manager, for windows I use choco, and for macos I use brew.


## MacOS
- Homebrew package manager installation guide, see [https://brew.sh/](https://brew.sh/)

```bash
brew upgrade
brew cask upgrade

brew cleanup
brew install git
brew install --cask visual-studio-code
brew install python
brew link python
brew install --cask spyder
python -m ensurepip --default-pip
pip install numpy pandas matplotlib pyside6 scikit-learn
```

## Windows 10/11
- choco package manager installation guide, see [https://chocolatey.org/install](https://chocolatey.org/install)
  
```bash
Set-ExecutionPolicy Unrestricted -Scope LocalMachine

choco upgrade all
c
choco install git
choco install vscode
choco install python
choco install spyder
python -m ensurepip --default-pip
pip install numpy pandas matplotlib pyside6 scikit-learn
```



In Windows you also need to go and disable program aliases


```zsh
Settings >> Apps >> Advanced App Settings >> App Execution Aliases

Disable:  App Installer assocaited with Python 

App Installer > python.exe
App Installer > python3.exe
```
