## Conditionals

### C
```c
int main(){
    if (3<10){
        // does somethinf
    }
    int n = 3;

    switch (n){
        case 1:
          // do something
          break;
        case 3:
          // do something else
          break;
        default:
          // otherwise
    }    
    return 0;
}
```

### Python

````python
# Conditionals in python (match, if)
print("If conditional:")
if 3<10:
    print("3 is less than 10.")
else:
    print("Not so much.")

print("Match conditional:")
val = 13
match val:
    case 10:
        print("The 10 Case.")
    case _:
        print("The Default case.")
````

## Loops

### C
```c
#include <stdio.h>
int main(){
    printf("For Loop:\n");
    for(int i = 0; i<5;i++){
        printf("%d\n",i);
    }

    int i = 0;
    printf("While Loop:\n");
    while (i<5){
        printf("%d\n",i);
        i++;
    }
}
```

### Python

````python
# Loops in python (for, while)

print("For Loop:")
for i in range(5):
    print(f"{i}")


print("While Loop:")
a = 0
while a < 5:
    print(f"{a}")
    a += 1
````

### Python uses Classes

````python

class MyCircle:
    def __init__(self, r, unit="cm"):
        self.r = r
        self.unit = unit

    # Method to calculate and print area of circle
    def calculate_area(self):
        print("Calculating area...")
        radius= self.r
        area = (radius)**2 * 3.14
        return area

    def print_area(self):
        print("Printing area...")
        radius = self.r
        area = self.calculate_area()
        print(f"Area of a Circle {radius=}{self.unit}: {area}{self.unit}^2")


circle = MyCircle(10.0)

circle.print_area()
    
````
